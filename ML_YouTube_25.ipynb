## <span style="font-size:25px">The backpropagation hidden layer formula</span>

|입력층 출력|은닉층 출력|
|:--------:|:--------:|
|$\text{A}1 = (\;\text{a}_1^{(1)}\quad\text{a}_2^{(1)})$|$\text{A}2 = (\;\text{a}_1^{(2)}\quad\text{a}_2^{(2)})$|

|출력층 손실|$\text{loss\_3} = \left( (\text{a}_1^{(3)} - \text{t}_1^{(3)})\text{a}_1^{(3)}(1 - \text{a}_1^{(3)})\qquad(\text{a}_2^{(3)} - \text{t}_2^{(3)})\text{a}_2^{(3)}(1 - \text{a}_2^{(3)}) \right)$|
|:--------:|:------------------------------------------------------:|
|은닉층 출력|$\text{loss\_2} = (\text{loss\_3} \bullet \text{W}3^{\text{T}}) \times (\text{A}2\times(1-\text{A}2))$|

$\frac {\partial\text{E}}{\partial\text{W}^{(2)}}$

$
\begin{bmatrix}

$

\begin{array}{c}

$
