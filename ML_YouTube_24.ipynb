## <span style="font-size:25px">The backpropagation formula for the output layer</span>
<span style="font-size:22px"> - Output layer weight</span>
$\text{W}^{(3)}$
<span style="font-size:22px"> / Output layer bias</span>
 $\text{b}^{(3)}$

<span style="font-size:20px">**[essential expression]**</span>
$i = (1, \:2)$

$
\text{E} = \text{E}_i + \text{E}_{i+1}
$

$
\text{E}_i = \frac 12 (\text{t}_i^{(3)} - \text{a}_i^{(3)})^2
$

$
\text{a}_i^{(3)} = \text{sigmiod}(\text{z}_i^{(3)})
$

$
\text{z}_i^{(3)} = \text{a}_i^{(2)}\text{W}_{i,\;i}^{(3)} + \text{a}_{i+1}^{(2)}\text{W}_{i,\;i+1}^{(3)} + \text{b}_i^{(3)}
$

## <span style="font-size:25px">The backpropagation output layer formula</span>

|은닉층 출력|$\text{A}2 = (\: \text{a}_1^{(2)} \quad \text{a}_2^{(2)})$|
|:--------:|:------------------------------------------------------:|
|출력층 손실|$\text{loss\_3} = \left( (\text{a}_1^{(3)} - \text{t}_1^{(3)})\text{a}_1^{(3)}(1 - \text{a}_1^{(3)})\qquad(\text{a}_2^{(3)} - \text{t}_2^{(3)})\text{a}_2^{(3)}(1 - \text{a}_2^{(3)}) \right)$|

<span style="font-size:22px">About </span>
$\frac {\partial\text{E}}{\partial\text{W}^{(3)}}$

$
\frac {\partial\text{E}}{\partial\text{W}^{(3)}}
$

=
$
\begin{bmatrix}
\frac {\partial\text{E}}{\partial\text{W}_{11}^{(3)}} &
\frac {\partial\text{E}}{\partial\text{W}_{12}^{(3)}} \\
\frac {\partial\text{E}}{\partial\text{W}_{21}^{(3)}} &
\frac {\partial\text{E}}{\partial\text{W}_{22}^{(3)}}
\end{bmatrix}
$

=
$
\begin{bmatrix}
(\text{a}_1^{(3)} - \text{t}_1^{(3)})\text{a}_1^{(3)}(1 - \text{a}_1^{(3)})\text{a}_1^{(2)} &
(\text{a}_2^{(3)} - \text{t}_2^{(3)})\text{a}_2^{(3)}(1 - \text{a}_2^{(3)})\text{a}_1^{(2)} \\
(\text{a}_1^{(3)} - \text{t}_1^{(3)})\text{a}_1^{(3)}(1 - \text{a}_1^{(3)})\text{a}_2^{(2)} &
(\text{a}_2^{(3)} - \text{t}_2^{(3)})\text{a}_2^{(3)}(1 - \text{a}_2^{(3)})\text{a}_2^{(2)}
\end{bmatrix}
$

=
$
\begin{bmatrix}
\text{a}_1^{(2)}(\text{a}_1^{(3)} - \text{t}_1^{(3)})\text{a}_1^{(3)}(1 - \text{a}_1^{(3)}) &
\text{a}_1^{(2)}(\text{a}_2^{(3)} - \text{t}_2^{(3)})\text{a}_2^{(3)}(1 - \text{a}_2^{(3)}) \\
\text{a}_2^{(2)}(\text{a}_1^{(3)} - \text{t}_1^{(3)})\text{a}_1^{(3)}(1 - \text{a}_1^{(3)}) &
\text{a}_2^{(2)}(\text{a}_2^{(3)} - \text{t}_2^{(3)})\text{a}_2^{(3)}(1 - \text{a}_2^{(3)})
\end{bmatrix}
$

=
$
\begin{bmatrix}
\text{a}_1^{(2)} \\ \text{a}_2^{(2)}
\end{bmatrix}
\bullet
\begin{bmatrix}
(\text{a}_1^{(3)} - \text{t}_1^{(3)})\text{a}_1^{(3)}(1 - \text{a}_1^{(3)})\qquad(\text{a}_2^{(3)} - \text{t}_2^{(3)})\text{a}_2^{(3)}(1 - \text{a}_2^{(3)})
\end{bmatrix}
$

=
$
\text{A}2^{\text{T}} \;\bullet\; \text{loss\_3}
$



<br><br><span style="font-size:22px">About </span>
$\frac {\partial\text{E}}{\partial\text{b}^{(3)}}$

=
$
\frac {\partial\text{E}}{\partial\text{b}^{(3)}}
$

=
$
\begin{bmatrix}
\frac {\partial\text{E}}{\partial\text{b}_1^{(3)}} & \frac {\partial\text{E}}{\partial\text{b}_2^{(3)}}
\end{bmatrix}
$

=
$
\begin{bmatrix}
(\text{a}_1^{(3)} - \text{t}_1^{(3)})\text{a}_1^{(3)}(1 - \text{a}_1^{(3)})\qquad(\text{a}_2^{(3)} - \text{t}_2^{(3)})\text{a}_2^{(3)}(1 - \text{a}_2^{(3)})
\end{bmatrix}
$

=
$
\text{loss\_3}
$

<br><br><span style="font-size:22px">- Update</span> $\text{W}^{(3)}$
<span style="font-size:22px"> output weights

$
\text{W}^{(3)} =
\text{W}^{(3)} - \alpha\frac{\partial\text{E}}{\partial\text{W}^{(3)}} =
\text{W}^{(3)} - \alpha \times (\text{A}2^{\text{T}} \bullet \text{loss\_3})
$

<br><span style="font-size:22px">- Update</span> $\text{b}^{(3)}$
<span style="font-size:22px"> output biases

$
\text{b}^{(3)} =
\text{b}^{(3)} - \alpha\frac{\partial\text{E}}{\partial\text{b}^{(3)}} = 
\text{b}^{(3)} - \alpha \times \text{loss\_3}
$
